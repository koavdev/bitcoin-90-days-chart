## Importing libraries

```python
import pandas as pd
from pycoingecko import CoinGeckoAPI
import plotly.offline as plotly
import plotly.graph_objs as go

plot = plotly.plot
```

## Initializing CoinGeckoAPI module

```python



cg = CoinGeckoAPI()
```

## Getting Bitcoin data

```python
bitcoin_data = cg.get_coin_market_chart_by_id(id="bitcoin", vs_currency="usd", days=90)
```

## Storing only price data

```python
bitcoin_price_data = bitcoin_data['prices']
```

## Creating a dataframe

```python
df = pd.DataFrame(bitcoin_price_data, columns=['TimeStamp', 'Price'])
```

```python



# Creating a new column as Date
df['Date'] = pd.to_datetime(df['TimeStamp'], unit='ms')

# Creating a candlestick with min, max, first and last prices
candlestick_data = df.groupby(df.Date.dt.date).agg({"Price": ["min", "max", "first" , "last"]})

# Creating the graph
fig = go.Figure(data=[go.Candlestick(x= candlestick_data.index,
                                   open = candlestick_data['Price']['first'],
                                   high = candlestick_data['Price']['max'],
                                   low = candlestick_data['Price']['min'],
                                   close = candlestick_data['Price']['last'])])

# Adding some titles
fig.update_layout(xaxis_rangeslider_visible=False, xaxis_title="Date", yaxis_title="Price (USD)", title="Bitcoin Chart Over Past 90 Days")

# Plot graph
fig.show()
```
